```python
import spacy
from datasets import load_from_disk
```

```python
dataset = load_from_disk("../data/processed/squad_knowledge_european_union_law")
```

```python
# 1. Cargar el modelo de inglés de spaCy
nlp = spacy.load("en_core_web_trf")
```

```python
list_texts = dataset['text'][:]
# ordenar los textos por longitud descendente
list_texts.sort(key=lambda x: len(x), reverse=True)
```

```python
text = list_texts[5]
print(text)
```

```python
doc = nlp(text)
```

```python
# 4. Imprimir todas las entidades encontradas
print("=== ENTIDADES (con tipo) ===")
print(len(doc.ents))
for ent in doc.ents:
    print(f"- {ent.text!r} -> {ent.label_}")
```

```python
unique_entities = sorted(set(ent.text for ent in doc.ents))
print("\n=== ENTIDADES ÚNICAS ===")
print(len(unique_entities))
for ent_text in unique_entities:
    print(f"- {ent_text}")
```
